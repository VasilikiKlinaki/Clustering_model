#Clustering

```python
#Create K-Means Clustering model with 2 clusters
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0)

kmeans.fit(X)
```

```python
kmeans.inertia_
```

```python
cols = X.columns
cols = X.columns
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()

X = ms.fit_transform(X)
X = pd.DataFrame(X, columns=[cols])
```

```python
y = dataset['status_type']
```

```python
kmeans.fit(X)
```

```python
kmeans.inertia_
```

```python
labels = kmeans.labels_


correct_labels = sum(y == labels)



print('Accuracy score: {}'.format(correct_labels / float(y.size)))
```

```python
cluster_results = []
```

```python
for i in range(1,11):

  kmeans = KMeans(n_clusters=i, random_state=0)
  kmeans.fit(X)
  cluster_results.append(kmeans.inertia_)

plt.plot(range(1,11), cluster_results)
plt.title("Elbow Method")
plt.xlabel("Number of Cluster")
plt.ylabel("Cluster_Results")
plt.show()
```

```python
#Create K-Means Clustering model with 3 clusters
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
```

```python
kmeans.inertia_
```

```python
labels = kmeans.labels_


correct_labels = sum(y == labels)



print('Accuracy score: {}'.format(correct_labels / float(y.size)))
```

```python
#Create K-Means Clustering model with 4 clusters
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
```

```python
# Plot features for visualization
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
labels= kmeans.labels_
```

```python
# pick two features to plot
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X['num_reactions'], X['num_comments'], c=labels, cmap="rainbow")
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, marker='X', label='centroids')
plt.title('Kmean clusters (Number of reactions vs Number of comments)')
plt.xlabel('Number of reaction')
plt.ylabel('Number of comments')
plt.show()
```
